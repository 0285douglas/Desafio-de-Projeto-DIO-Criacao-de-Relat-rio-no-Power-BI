```python
import matplotlib.pyplot as plt
import seaborn as sns
```

Identificação dos Itens mais vendidos


```python
plt.figure(figsize=(30,10))
sns.countplot(x='Item', data=mais_vendidos, palette='Set2')

plt.title('Gráfico de Contagem por Categoria', fontsize=16)
plt.xlabel('Categorias', fontsize=14)
plt.ylabel('Contagem', fontsize=14)

plt.show()
```

```python
mais_vendidos['Item'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title('Itens mais vendidos')
plt.ylabel('')
plt.show()
```
